```python
from langchain_community.chat_models import ChatOllama
from langchain.schema import HumanMessage
```

```python
llm = ChatOllama(
    model="llama3.1",
    temperature=0.6,
    # other params...
)
```

```python
messages = [HumanMessage(content="What is max output of llama 3.1")]
name = llm(messages)
print(name.content)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
prompt_template_name = PromptTemplate(
    input_variables=['cuisine'],
    template="I want to open a restaurant with {cuisine} cuisine. Suggest a fancy name for this.",

)
prompt_template_name.format(cuisine="Italian")


name_chain = LLMChain(llm=llm, prompt=prompt_template_name)


prompt_template_items = PromptTemplate(
    input_variables=['restaurant_name'],
    template="Suggest some menu items for {restaurant_name}, Return a comma seperated list"
)

food_items_chain = LLMChain(llm = llm, prompt = prompt_template_items)
```

```python
from langchain.chains import SimpleSequentialChain
chain = SimpleSequentialChain(chains=[name_chain, food_items_chain])
result = chain.run('Indian')
```

```python
print(f"Result: {result}")
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
prompt_template_name = PromptTemplate(
    input_variables=['cuisine'],
    template="Give straight answer directly. I want to open a restaurant with {cuisine} cuisine. Suggest one fancy name for this.",

)
prompt_template_name.format(cuisine="Italian")


name_chain = LLMChain(llm=llm, prompt=prompt_template_name, output_key = 'restaurant_name')


prompt_template_items = PromptTemplate(
    input_variables=['restaurant_name'],
    template="Suggest 10 menu items for {restaurant_name}, Return a comma seperated list"
)

food_items_chain = LLMChain(llm = llm, prompt = prompt_template_items, output_key='menu_items')
```

```python
from langchain.chains import SequentialChain
chain = SequentialChain(chains=[name_chain, food_items_chain], input_variables=["cuisine"], output_variables = ['restaurant_name', 'menu_items'])
result = chain({'cuisine': 'arabic'})
print(f'{result}')
```

```python
from langchain.memory import ConversationBufferMemory
memory = ConversationBufferMemory()
chain = LLMChain(llm=llm, prompt=prompt_template_name, memory = memory)
name = chain.run("mexican")
print(f"{name}")
```

```python
name = chain.run("Indian")
print(f"{name}")
```

```python
print(chain.memory.buffer)
```
